```python
import math
import numpy as np 
import matplotlib.pyplot as plt

size_1 = 5

n = np.arange(0, size_1, 0.1)
plt.plot(n, (2**20) * n + 2 , 'red', n, n**7.1 + 2.1**20, 'blue', n, 4**n - 2.1**8, 'green')
plt.xlim(0, size_1)
plt.rcParams["figure.figsize"] = (7,7)
plt.show()
```
